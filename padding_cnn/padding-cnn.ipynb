```python
# Padding in keras - valid and same 
# Author: Muhammad Humayun Khan

import tensorflow
from tensorflow import keras
from tensorflow.keras.layers import Dense, InputLayer,Conv2D, Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import mnist
```

```python
# load data from the dataset mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

# **Not applying padding - valid**

```python
# build the model
model = Sequential()

model.add(InputLayer(shape = (28,28,1,)))   # 28x28 image
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu')) # Conv2D is like Dense having 32 filters
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu')) # kernel_size is the size of the filter 3x3
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))

model.add(Flatten())  # convert the data into the 1D

# now send the data to the normal neural architecture structure
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
# after first layer from 28x28 into 26x26, then 24x24, then 22x22 then flatten
```

# Applying padding - same

```python
# build the model
model = Sequential()

model.add(InputLayer(shape = (28,28,1,)))   # 28x28 image
model.add(Conv2D(32,kernel_size=(3,3),padding='same', activation='relu')) # Conv2D is like Dense having 32 filters
model.add(Conv2D(32,kernel_size=(3,3),padding='same', activation='relu')) # kernel_size is the size of the filter 3x3
model.add(Conv2D(32,kernel_size=(3,3),padding='same', activation='relu'))

model.add(Flatten())  # convert the data into the 1D

# now send the data to the normal neural architecture structure
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()

# now the value will remain as 28x28
```
