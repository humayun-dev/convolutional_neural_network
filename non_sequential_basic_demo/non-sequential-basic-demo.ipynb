```python
# Non-sequential Neural Network implementation
# The yearly_salary and height will be the input columns and to determine the age and city belonging of a person
# Author: Muhammad Humayun Khan

from keras.models import Model
from keras.layers import *
```

```python
# define the inputs and outputs to the model
model = Model(inputs = x,outputs = [output_age,output_place])
```

```python
# input layer
x = Input(shape=(3,))
```

```python
# hidden layers
hidden_one = Dense(128,activation = 'relu')(x)   # (x) is the input attached to the layer
hidden_two = Dense(64,activation = 'relu')(hidden_one)  # (hidden_one) is the input attached to the layer
```

```python
# branching where we have to predict the output
output_age = Dense(1,activation = 'linear')(hidden_two)  # output_age concerns with the hidden_two
output_place = Dense(1,activation = 'sigmoid')(hidden_two) # output_place concers with the hidden_two
```
