```python
# Data augmentation implementation - using one image to generate many
# Author: Muhammad Humayun Khan

import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = '/content/drive/MyDrive/images/cat.jpg'
img = image.load_img(image_path,target_size = (200,200))
```

```python
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 0.2,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    vertical_flip = True
)
```

```python
img = image.img_to_array(img)

```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i = 0
for output in datagen.flow(input_batch,batch_size = 1,save_to_dir = '/content/drive/MyDrive/images/',save_prefix = 'cat',save_format = 'jpeg'):
  i = i + 1
  if i == 10:
    break
```
